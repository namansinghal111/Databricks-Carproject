### **DEEP CLONE**

```python
%sql
CREATE TABLE databricksansh.silver.deepclonenew
CLONE databricksansh.silver.sales_enr
```

```python
%sql
DESCRIBE HISTORY databricksansh.silver.deepclonenew
```

```python
%sql
SELECT * FROM databricksansh.silver.deepclonenew
```

### **SHALLOW CLONE**

```python
%sql
CREATE TABLE databricksansh.silver.shallow
SHALLOW CLONE databricksansh.silver.sales_enr
```
